```python
def filter_cars(catalogue, parameter, criterion):
    desired_cars = []
    for car_info_ in catalogue:
        car_, car_info_ = list(car_info_.items())[0]
        if car_info_[parameter] == criterion:
            desired_cars.append(car)
    return desired_cars
```

```python
cars_catalog = []
for car, car_info in cars.items():
    cars_catalog.append({car:car_info})
```

```python
print(filter_cars(cars_catalog, 'Цена', 50000))
```
